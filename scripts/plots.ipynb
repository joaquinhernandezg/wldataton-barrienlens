```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model_name = "hyperunet_wmse"

log = pd.read_csv(f"/storage/jhernand.2024/dataton/wl_dataton/models/{model_name}/{model_name}_training_log.log")
# plot val_los, val]
fig, axs = plt.subplots(1, 4, figsize=(20, 4), sharex=True)
fig.suptitle("UNET WMSE")
axs[0].plot(log["loss"], label="loss")
axs[0].plot(log["val_loss"], label="val_loss")
axs[0].legend()


axs[1].plot(log["dpeaks"], label="dpeaks")
axs[1].plot(log["val_dpeaks"], label="val_dpeaks")
axs[1].legend()


axs[2].plot(log["wmape"], label="wmape")
axs[2].plot(log["val_wmape"], label="val_wmape")
axs[2].legend()


axs[3].plot(log["dicee"], label="dicee")
axs[3].plot(log["val_dicee"], label="val_dicee")
axs[3].legend()
axs[0].set_ylim(0, 1)
axs[1].set_ylim(10, 60)
axs[2].set_ylim(0, 3)
axs[3].set_ylim(0, 0.1)


model_name = "hyperunet"

log = pd.read_csv(f"/storage/jhernand.2024/dataton/wl_dataton/models/{model_name}/{model_name}_training_log_new.log")
# plot val_los, val]
fig, axs = plt.subplots(1, 4, figsize=(20, 4), sharex=True)
fig.suptitle("UNET MULTIFOCAL LOSS")

axs[0].set_xlim(0, 100)
axs[0].plot(log["loss"], label="loss")
axs[0].plot(log["val_loss"], label="val_loss")

axs[0].legend()


axs[1].plot(log["dpeaks"], label="dpeaks")
axs[1].plot(log["val_dpeaks"], label="val_dpeaks")

axs[1].legend()


axs[2].plot(log["wmape"], label="wmape")
axs[2].plot(log["val_wmape"], label="val_wmape")


axs[2].legend()


axs[3].plot(log["dicee"], label="dicee")
axs[3].plot(log["val_dicee"], label="val_dicee")
axs[0].set_ylim(0, 1)
axs[1].set_ylim(10, 60)
axs[2].set_ylim(0, 3)
axs[3].set_ylim(0, 0.1)

axs[3].legend()
```
